```python
import pandas as pd  # load and manipulate data and for One-Hot Encoding
import numpy as np  # calculate the mean and standard deviation
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb  # XGBoost stuff
from sklearn.model_selection import train_test_split  # split data into training and testing sets
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer  # for scoring during
from sklearn.model_selection import GridSearchCV  # cross validation
from sklearn.metrics import confusion_matrix  # creates a confusion matrix
from sklearn.metrics import ConfusionMatrixDisplay  # draws a confusion matrix
from sklearn.preprocessing import OneHotEncoder  # Más explícito para encoding
from sklearn.metrics import classification_report

from sklearn.preprocessing import LabelEncoder, StandardScaler
from xgboost import XGBRegressor
from xgboost import XGBClassifier
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import RandomizedSearchCV
from xgboost import plot_importance
```

Entrenamiento del Modelo

```python
X_encoded = pd.read_csv('X_encoded.csv', encoding='utf-8', low_memory=False)
y = pd.read_csv('y.csv', encoding='utf-8', low_memory=False)
```

```python
#Dividir datos
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

y_train = np.squeeze(y_train)
y_test = np.squeeze(y_test)

#Estructura DMatrix
dtrain = xgb.DMatrix(X_train, label=y_train)
dtest = xgb.DMatrix(X_test, label=y_test)

print("X shape:", X_encoded.shape)
print("y shape:", y.shape)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)

print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Distribución de y_train vs y_test
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(y_train, bins=30, color='blue', alpha=0.7)
plt.title('Distribución de y_train')
plt.subplot(1, 2, 2)
plt.hist(y_test, bins=30, color='red', alpha=0.7)
plt.title('Distribución de y_test')
plt.show()
```

## HIPERPARAMETRIZACION MODELO A

```python
# Espacio de búsqueda de hiperparámetros
param_dist = {
    'learning_rate': np.linspace(0.01, 0.3, 10),
    'max_depth': range(3, 10),
    'n_estimators': range(50, 300, 50),
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.2],
}

model = XGBRegressor(
    objective='reg:squarederror',  # Error cuadrático para regresión
    enable_categorical=True,       # Para tus columnas tipo 'category'
    tree_method='hist',            # Eficiente con variables categóricas
    n_estimators=500,             # Número de árboles (aumenta si es necesario)
)

# Búsqueda aleatoria con validación cruzada
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_dist,
    n_iter=50,  # Número de combinaciones a probar
    cv=5,       # Validación cruzada de 5 folds
    scoring='accuracy',  # o 'neg_mean_squared_error' para regresión
    verbose=2,
    random_state=42
)

random_search.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=1)

# Mejores parámetros
print("Mejores parámetros:", random_search.best_params_)
print("Mejor score:", random_search.best_score_)
```

## HIPERPARAMETRIZACION MODELO B

Segundo intento de Hiperparametrizacion

```python
model = XGBRegressor(
    tree_method='gpu_hist',
    predictor='gpu_predictor',
    gpu_id=0,
    objective='reg:squarederror',
    eval_metric='rmse',
    verbosity=1
)

param_grid = {
    'max_depth': [3, 5, 7],              # más profundidad = más complejidad
    'learning_rate': [0.01, 0.05, 0.1],  # evita valores muy altos como 0.3 al inicio
    'subsample': [0.7, 0.9],             # regularización natural
    'colsample_bytree': [0.7, 0.9],      # igual, para evitar overfitting
    'n_estimators': [100, 200, 300],     # más = más potencia, pero más tiempo
    'gamma': [0, 1],                     # para controlar el crecimiento de los árboles
    'reg_lambda': [1, 10],               # regularización L2
    'reg_alpha': [0, 1]                  # regularización L1
}

random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_grid,
    n_iter=25,  # Mucho más rápido
    scoring='neg_root_mean_squared_error',
    cv=3,
    verbose=1,
    n_jobs=1
)

random_search.fit(
    X_train, y_train,
    eval_set=[(X_test, y_test)],
    verbose=1)
```

Resultado

```python
print("Best parameters:", random_search.best_params_)
print("Best RMSE (neg):", random_search.best_score_)
```

```python
# Entrenar modelo a
params_a = { 
    "learning_rate": 0.3,
    "max_depth": 4,
    "gamma": 0,
    "subsample": 0.8,
    "colsample_bytree": 1.0,
    "objective": "reg:squarederror",
    "tree_method": "gpu_hist",
    "predictor": "gpu_predictor",
    "gpu_id": 0,
    "eval_metric": "rmse"
}

model_a = xgb.train(
    params=params_a,
    dtrain=dtrain,
    num_boost_round=200,
    evals=[(dtest, "Test"), (dtrain, "Train")],
    early_stopping_rounds=10
)
```

```python
# Entrenar modelo b
params_b = {
    "reg_lambda": 1,
    "reg_alpha": 0,
    "learning_rate": 0.1,
    "max_depth": 7,
    "gamma": 0,
    "subsample": 0.9,
    "colsample_bytree": 0.9,
    "objective": "reg:squarederror",
    "tree_method": "gpu_hist",
    "predictor": "gpu_predictor",
    "gpu_id": 0,
    "eval_metric": "rmse"
}

model_b = xgb.train(
    params=params_b,
    dtrain=dtrain,
    num_boost_round=200,
    evals=[(dtest, "Test"), (dtrain, "Train")],
    early_stopping_rounds=10
)
```

```python
# Entrenar modelo c
params_c = {
    "reg_lambda": 1,
    "reg_alpha": 0,
    "learning_rate": 0.05,
    "max_depth": 7,
    "gamma": 0,
    "subsample": 0.9,
    "colsample_bytree": 0.9,
    "objective": "reg:squarederror",
    "tree_method": "gpu_hist",
    "predictor": "gpu_predictor",
    "gpu_id": 0,
    "eval_metric": "rmse"
}

model_c = xgb.train(
    params=params_c,
    dtrain=dtrain,
    num_boost_round=400,
    evals=[(dtest, "Test"), (dtrain, "Train")],
    early_stopping_rounds=10
)
```

```python

# Modelo A
y_pred_a = model_a.predict(dtest)
rmse_a = mean_squared_error(y_test, y_pred_a, squared=False)
r2_a = r2_score(y_test, y_pred_a)

# Modelo B
y_pred_b = model_b.predict(dtest)
rmse_b = mean_squared_error(y_test, y_pred_b, squared=False)
r2_b = r2_score(y_test, y_pred_b)

# Modelo C
y_pred_c = model_c.predict(dtest)
rmse_c = mean_squared_error(y_test, y_pred_c, squared=False)
r2_c = r2_score(y_test, y_pred_c)

print("Modelo A - RMSE:", rmse_a, "R²:", r2_a)
print("Modelo B - RMSE:", rmse_b, "R²:", r2_b)
print("Modelo C - RMSE:", rmse_c, "R²:", r2_c)
```

```python
df_comparacion = pd.DataFrame({
    'Modelo': ['A', 'B', 'C'],
    'RMSE': [rmse_a, rmse_b, rmse_c],
    'R2': [r2_a, r2_b, r2_b]
})

print(df_comparacion)
```
