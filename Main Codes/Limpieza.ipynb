Para conocer la distribuci贸n de los tipos de atenci贸n registrados, se construy贸 una tabla de frecuencias que muestra cu谩ntas veces aparece cada categor铆a en la base de datos. Esta informaci贸n es clave para entender qu茅 modalidades concentran la mayor demanda de servicios.

Podemos revisar si `Nombre_Tipo_Atencion_Arp` aparece en los valores 煤nicos con el siguiente c贸digo:

```python
df8['Nombre_Tipo_Atencion_Arp'].unique()
```

La categor铆a 'Nombre_Tipo_Atencion_Arp' no deber铆a estar como un nivel de la variable. Eso indica que en alg煤n momento se col贸 una fila del encabezado como si fuera un dato. Adem谩s, tambi茅n hay un `NaN`, lo cual sugiere que hay registros vac铆os en esa columna. Lo ideal es limpiar esos valores incorrectos antes de generar las tablas o gr谩ficos. Este c贸digo asegurar谩 que solo aparezcan las categor铆as v谩lidas, eliminando tanto los `NaN` como el valor incorrecto que parece haberse filtrado desde el encabezado.

```python
# Eliminar filas con valores incorrectos o vac铆os en 'Nombre_Tipo_Atencion_Arp'
df9 = df8[
    (df8['Nombre_Tipo_Atencion_Arp'].notna()) &
    (df8['Nombre_Tipo_Atencion_Arp'] != 'Nombre_Tipo_Atencion_Arp')
]
```

Luego, podemos volver a correr la tabla de frecuencias:

```python
# Tabla de frecuencias con proporciones
frecuencias = df9['Nombre_Tipo_Atencion_Arp'].value_counts(normalize=False).reset_index()
frecuencias.columns = ['Tipo de Atenci贸n', 'Frecuencia']

# Calcular proporci贸n
frecuencias['Proporci贸n (%)'] = (frecuencias['Frecuencia'] / frecuencias['Frecuencia'].sum()) * 100
frecuencias['Proporci贸n (%)'] = frecuencias['Proporci贸n (%)'].round(2)

# Mostrar la tabla
print(frecuencias)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
valores = frecuencias['Frecuencia']
etiquetas = frecuencias['Tipo_Llamada_Arp_Desc']
porcentajes = (valores / valores.sum()) * 100
porcentajes_redondeados = porcentajes.round(1)

# Leyenda: nombre (porcentaje%)
etiquetas_legenda = [f"{etiqueta} ({pct:.1f}%)" for etiqueta, pct in zip(etiquetas, porcentajes_redondeados)]

# Crear figura
fig, ax = plt.subplots(figsize=(10, 8))

# Pie chart sin etiquetas (se colocar谩n manualmente los porcentajes grandes)
wedges, _ = ax.pie(valores, startangle=90, radius=1.0)

# Anotar porcentajes solo si son >= 1%
for i, (wedge, pct) in enumerate(zip(wedges, porcentajes)):
    if pct >= 1:
        ang = (wedge.theta2 + wedge.theta1) / 2
        x = np.cos(np.deg2rad(ang)) * 0.6
        y = np.sin(np.deg2rad(ang)) * 0.6
        ax.text(x, y, f"{pct:.1f}%", ha='center', va='center', fontsize=9)

# Leyenda con nombre y porcentaje
ax.legend(wedges, etiquetas_legenda, title="Tipo de Llamada", loc="center left", bbox_to_anchor=(1, 0.5), fontsize=9)

# T铆tulo y ajustes finales
ax.set_title("Distribuci贸n porcentual por tipo de llamada", fontsize=14)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
